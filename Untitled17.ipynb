```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('Unemployment in India.xlsx')
df.head()
print(df.shape)
print(df.columns)
df.info()
df.columns = [
    'State',
    'Date',
    'Frequency',
    'Estimated_Unemployment_Rate',
    'Estimated_Employed',
    'Estimated_Labour_Participation_Rate',
    'Region'
]
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df.isnull().sum()
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated_Unemployment_Rate', data=df)
plt.title('Unemployment Rate in India Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
state_unemployment = df.groupby('State')['Estimated_Unemployment_Rate'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
state_unemployment.plot(kind='bar')
plt.title('Average Unemployment Rate by State')
plt.ylabel('Unemployment Rate (%)')
plt.show()
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Estimated_Unemployment_Rate', data=df)
plt.title('Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.show()
covid_df = df[df['Date'] >= '2020-03-01']

plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated_Unemployment_Rate', data=covid_df)
plt.title('Unemployment Rate During Covid-19')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
print("Highest Unemployment Rate:", df['Estimated_Unemployment_Rate'].max())
print("Lowest Unemployment Rate:", df['Estimated_Unemployment_Rate'].min())
```
